#ðŸ“Œ ExtracciÃ³n

importar los datos de la API de Telecom X. Estos datos estÃ¡n disponibles en formato JSON junto a las bibliotecas a utilizar

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json'
df = pd.read_json(url)
df.head()
```

```python
df.tail()
```

```python
df.sample(7)
```

```python
df.info()
```

#ðŸ”§ TransformaciÃ³n

```python
df_normalizado = pd.json_normalize(
    data=df.to_dict(orient='records'),
    record_path=None,
    meta=['customerID', 'Churn'],
    errors='ignore'
)

nuevos_nombres = {
    'customer.gender': 'genero del cliente',
    'customer.SeniorCitizen': 'cliente senior',
    'customer.Partner': 'cliente con pareja',
    'customer.Dependents': 'cliente con dependientes',
    'phone.PhoneService': 'cliente con servicio telefonico',
    'phone.MultipleLines': 'cliente con multiples lineas',
    'internet.InternetService': 'servicio de internet',
    'internet.OnlineSecurity': 'seguridad en linea',
    'internet.OnlineBackup': 'backup en linea',
    'internet.DeviceProtection': 'proteccion de dispositivo',
    'internet.TechSupport': 'soporte tecnico',
    'internet.StreamingTV': 'streaming tv',
    'internet.StreamingMovies': 'streaming movies',
    'account.Contract': 'Tipo de contrato',
    'account.PaperlessBilling': 'Facturacion en linea',
    'account.PaymentMethod': 'Metodo de pago',
    'account.Charges.Monthly': 'costo mensual',
    'account.Charges.Total': 'costo total',
    'customer.tenure':'Meses de contrato del cliente'
}

df_normalizado = df_normalizado.rename(columns=nuevos_nombres)
df_normalizado.head()
```

```python
df_normalizado.info()
```

```python
df_normalizado['costo total']=df_normalizado['costo total'].replace(' ',0.0)
df_normalizado['costo total'] = df_normalizado['costo total'].astype(np.float64)
df_normalizado.info()
```

```python
df_normalizado.isna().isnull().sum()
```

```python
df_normalizado['Churn'] = df_normalizado['Churn'].replace('','unknown')
df_normalizado['Churn'].unique()
```

```python
for i in df_normalizado.select_dtypes(include='object').columns:
  df_normalizado[i]=df_normalizado[i].str.lower().str.strip()
```

```python
df_normalizado['costos diarios']=df_normalizado['costo mensual']/30
df_normalizado.head()
```

#ðŸ“Š Carga y anÃ¡lisis

####calculo de la media, mediana, moda, desviacion estandar y medidas de distribucion y comportamiento

```python
df_normalizado_medidas = df_normalizado.describe().round(2)
df_normalizado_medidas
```

####grÃ¡ficos para visualizar la proporciÃ³n de clientes que permanecieron y los que se dieron de baja usando la variable 'churn'.

```python
# GrÃ¡fico de barras para la proporciÃ³n de clientes que permanecieron y los que se dieron de baja
churn_counts = df_normalizado['Churn'].value_counts().reset_index()
churn_counts.columns = ['Churn', 'Count']

plt.figure(figsize=(6, 4))
sns.barplot(x='Churn', y='Count', data=churn_counts, palette='Set2')
plt.title('ProporciÃ³n de clientes por Churn')
plt.xlabel('Churn')
plt.ylabel('NÃºmero de clientes')
plt.show()
```

####distribucion de evasiÃ³n segÃºn variables categÃ³ricas, como gÃ©nero, tipo de contrato, mÃ©todo de pago, entre otras.*texto en cursiva*

```python
# DistribuciÃ³n de evasiÃ³n segÃºn variables categÃ³ricas
categorical_cols = ['genero del cliente', 'cliente senior', 'cliente con pareja', 'cliente con dependientes',
                    'cliente con servicio telefonico', 'cliente con multiples lineas', 'servicio de internet',
                    'seguridad en linea', 'backup en linea', 'proteccion de dispositivo', 'soporte tecnico',
                    'streaming tv', 'streaming movies', 'Tipo de contrato', 'Facturacion en linea', 'Metodo de pago']

plt.figure(figsize=(10, 40))
plt.subplots_adjust(hspace=0.4, wspace=0.3)
for i, col in enumerate(categorical_cols):
    plt.subplot(8, 2, i + 1)
    sns.countplot(x=col, hue='Churn', data=df_normalizado)
    plt.title(f'DistribuciÃ³n de Churn por {col}')
    plt.xlabel(col)
    plt.ylabel('NÃºmero de clientes')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

####exploracion de las variables numÃ©ricas, como "total gastado" o "tiempo de contrato", y como se distribuyen entre los clientes que cancelaron y los que no cancelaron por medio de graficos.

```python
numerical_cols = ['Meses de contrato del cliente', 'costo mensual', 'costo total', 'costos diarios']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=df_normalizado, x=col, hue='Churn', kde=True)
    plt.title(f'DistribuciÃ³n de {col} por Churn')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 20))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(data=df_normalizado, x='Churn', y=col, hue='Churn', palette='Set2', legend=False) # Added hue='Churn' and legend=False
    plt.title(f'DistribuciÃ³n de {col} por Churn (Box Plot)')
    plt.xlabel('Churn')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```
